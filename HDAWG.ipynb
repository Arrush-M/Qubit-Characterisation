```python
from zhinst.qcodes import HDAWG
from scipy import signal
import numpy as np
```

```python
# Preliminary, to check connection. No need to run this every time.
hdawg = HDAWG("DEV8488", "localhost")
hdawg.awgs[0].set_sequence_params(
    sequence_type="Simple",
    period=20e-6,
    repetitions=1000
)

w_cosine = np.array([np.cos(2 * np.pi * 1e6 * t) for t in np.linspace(0, 20e-6, 8000)])
hdawg.awgs[0].reset_queue()
hdawg.awgs[0].queue_waveform(w_cosine, [])
hdawg.awgs[0].compile_and_upload_waveforms()
hdawg.sigouts[0].on(True)
hdawg.awgs[0].enable(True)
```
